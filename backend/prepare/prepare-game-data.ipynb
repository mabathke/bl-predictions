# Prepare Game Data

```python
import pandas as pd
import os 
```

```python
def clean_results(df: pd.DataFrame, columns_to_drop: list, isFirstLiga: bool) -> pd.DataFrame:
    # Split 'Spielpaarung' into home and away teams (and strip any extra whitespace)
    df["HeimTeam"] = df["Spielpaarung"].str.split("-").str[0].str.strip()
    df["GastTeam"] = df["Spielpaarung"].str.split("-").str[1].str.strip()
    
    # Convert the scores to integers for accurate comparisons
    heim_goals = df["Ergebnis"].str.split(":").str[0].astype(int)
    gast_goals = df["Ergebnis"].str.split(":").str[1].astype(int)
    
    # Determine wins based on numerical comparison
    df["HeimSieg"] = heim_goals > gast_goals
    df["GastSieg"] = heim_goals < gast_goals

    # Save the number of goals as integers
    df["HeimTore"] = heim_goals
    df["GastTore"] = gast_goals
    
    # 
    df["ErsteLiga"] = isFirstLiga
    
    # Drop the specified columns; columns_to_drop should be a list of column names
    df.drop(columns=columns_to_drop, inplace=True)
    
    return df
```

## Prepare the data for results of the first bl

```python
cwd = os.getcwd()

# Construct the correct relative path
csv_path = os.path.join(cwd, "../gather/data/first/first_bl_results.csv")
print("CSV path:", csv_path)

# Load the CSV file with pandas
first_df = pd.read_csv(csv_path)
first_df
```

```python
colonns_to_drop = ["Unnamed: 0", "Spielpaarung", "Ergebnis"]
first_df = clean_results(first_df, colonns_to_drop, True)
first_df
```

## Prepare Second Bundesliga Data

```python
cwd = os.getcwd()

# Construct the correct relative path
csv_path = os.path.join(cwd, "../gather/data/second/second_bl_results.csv")
print("CSV path:", csv_path)

# Load the CSV file with pandas
second_df = pd.read_csv(csv_path)
second_df
```

```python
colonns_to_drop = ["Spielpaarung", "Ergebnis"]
second_df = clean_results(second_df, colonns_to_drop, False)
second_df
```

Now, lets combine the data from the first and second Bundesliga and prepare it for analysis. We have some team names that differ but represent the same team. We will fix this by mapping the names to the same name.

```python
all_results_df = pd.concat([first_df, second_df])
```

```python
from Levenshtein import distance

# Assume all_results_df is your DataFrame
# Get the unique team names from both columns
home_teams = all_results_df['HeimTeam'].unique()
away_teams = all_results_df['GastTeam'].unique()

# Combine and get unique team names overall
teams = pd.unique(list(home_teams) + list(away_teams))

# List to store pairs with a max edit distance of 4
similar_pairs = []

# Compare each pair of teams (without repetition)
for i in range(len(teams)):
    for j in range(i+1, len(teams)):
        dist = distance(teams[i], teams[j])
        if dist <= 3:
            similar_pairs.append((teams[i], teams[j], dist))

# Display the similar pairs
print("Team name pairs differing by 4 characters or less:")
for team1, team2, dist in similar_pairs:
    print(f"{team1}  <-->  {team2}  (Edit Distance: {dist})")
```

```python
manual_mappings = [
    ("Werder Bremen", "SV Werder Bremen"),
    ("Schalke 04", "FC Schalke 04"),
    ("Hannover 96", "Hannover 1896"),
    ("Bayern München", "FC Bayern München"),
    ("Hansa Rostock", "FC Hansa Rostock"),
    ("Bayer Uerdingen", "Bayer 05 Uerdingen"),
    ("Darmstadt 98", "SV Darmstadt 98"),
    ("Bayer Leverkusen", "Bayer 04 Leverkusen"),
    ("Wattenscheid 09", "SG Wattenscheid 09"),
    ("Dynamo Dresden", "SG Dynamo Dresden"),
    ("KFC Uerdingen", "KFC Uerdingen 05"),
    ("SpVgg Unterhaching", "SpVgg. Unterhaching"),
    ("Energie Cottbus", "FC Energie Cottbus"),
    ("FSV Mainz 05", "1. FSV Mainz 05"),
    ("SpVgg Greuther Fürth", "SpVgg. Greuther Fürth"),
    ("SC Paderborn 07", "SC Paderborn"),
    ("FC Erzgebirge Aue", "Erzgebirge Aue"),
]

# 2. Create a dictionary mapping variant -> canonical name
mapping_dict = {variant: canonical for canonical, variant in manual_mappings}

# 3. Apply the mapping to both HeimTeam and GastTeam columns:
all_results_df['HeimTeam'] = all_results_df['HeimTeam'].replace(mapping_dict)
all_results_df['GastTeam'] = all_results_df['GastTeam'].replace(mapping_dict)
```
